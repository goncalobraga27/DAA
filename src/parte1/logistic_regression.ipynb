```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score, train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, accuracy_score
```

```python
df = pd.read_csv('../../datasets/parte1/dataset_cleaned1.csv')
```

```python
X = df.drop(['EncodedRainTomorrow'], axis=1)
y = df['EncodedRainTomorrow'].to_frame()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=2023)
```

```python
logmodel1 = LogisticRegression(random_state=2023, solver='newton-cg')
```

```python
logmodel1.fit(X_train, y_train)
```

```python
predictions1 = logmodel1.predict(X_test)
```

```python
print("%0.2f accuracy" % (accuracy_score(y_test, predictions1)))

print(classification_report(y_test,predictions1))

ConfusionMatrixDisplay.from_predictions(y_test, predictions1)
plt.show()
```
