```python
fig = plt.figure(figsize=(10, 10))
satisfaction_corr = df.corr(method='pearson', numeric_only=True)
sns.heatmap(satisfaction_corr, linecolor='black', linewidths=0.5, annot=True)
```

```python
satisfaction_count = df['satisfaction'].value_counts()
sns.set_style('darkgrid')
sns.barplot(x=satisfaction_count.index, y=satisfaction_count.values)
plt.title('Frequency Distribution of Satisfaction')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Satisfaction', fontsize=12)
plt.show()
```

```python
labels = df['satisfaction'].astype('category').cat.categories.tolist()
counts = df['satisfaction'].value_counts()
sizes = [counts[var_cat] for var_cat in labels]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True)
ax1.axis('equal')
plt.show()
```

```python
fig, axs = plt.subplots(1, 1, figsize=(12, 10))
fig.suptitle('Departure Delay in Minutes Boxplot')
sns.boxplot(y=df['Departure Delay in Minutes'], ax=axs)
axs.set_ylim(-100, 200) 
```

```python
fig, axs = plt.subplots(1, 1, figsize=(12, 10))
fig.suptitle('Arrival Delay in Minutes Boxplot')
sns.boxplot(y=df['Arrival Delay in Minutes'], ax=axs)
axs.set_ylim(-100, 200) 
```
