# Exploração de dados do dataset Global de Treino 

### Import Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import spearmanr
%matplotlib inline
```

### Get the data 

```python
df = pd.read_csv('../../datasets/parte2/treino/datasetGlobal.csv')
```

### Correlation beetween features 

```python
fig = plt.figure(figsize=(10, 10))
corr = df.corr(method='pearson', numeric_only=True)
sns.heatmap(corr, linecolor='black', linewidths=0.5, annot=True)
```
